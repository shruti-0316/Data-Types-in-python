

*   **Byte objects vs string in python**


```python
#  String encoding

# initialising a String
a = 'Hello'

# initialising a byte object
c = b'Hello'

# using ASCII mapping
d = a.encode('ASCII')

# checking if a is converted to bytes or not
if (d==c):
	print ("Encoding successful")
else : print ("Encoding Unsuccessful")
```
